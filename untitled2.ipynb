#Ön İşleme AşamalarıGenel Temizlik

```python
import re

# Genel temizlik fonksiyonu
def temizle(text):
    text = re.sub(r"<.*?>", " ", text)        # HTML etiketlerini sil
    text = re.sub(r"[^a-zA-ZçğıöşüÇĞİÖŞÜ\s]", " ", text)  # Özel karakterleri sil
    text = re.sub(r"\s+", " ", text)          # Fazla boşlukları sil
    return text.strip()

df["urun_temiz"] = df["urun_lower"].apply(temizle)
display(df[["urun_lower", "urun_temiz"]].head())
```

# Lemmatization

```python
import pandas as pd
import ast
from zemberek.morphology import TurkishMorphology

# Initialize the Zemberek Turkish Morphology object for lemmatization
# Ensure that Java is correctly configured in the environment for this to work.
morphology = TurkishMorphology.create_with_defaults()

# Function to apply Zemberek lemmatization
def apply_zemberek_lemmatization(tokens):
    if not isinstance(tokens, list):
        # Try to convert string representation of list to actual list
        try:
            tokens = ast.literal_eval(tokens)
        except (ValueError, SyntaxError):
            return ["ERROR_ZEMBEREK_LEMMA_PARSE"]

    if not tokens:
        return []

    lemmas = []
    for token in tokens:
        # morphology.analyze(token) returns a WordAnalysis object (a wrapper for Java's WordAnalysis).
        # This WordAnalysis object *contains* a list of SingleWordAnalysis objects.
        # Assuming this list is exposed via an attribute named 'analyses'.
        word_analysis_obj = morphology.analyze(token)

        if word_analysis_obj and hasattr(word_analysis_obj, 'analyses') and word_analysis_obj.analyses:
            # Access the list of SingleWordAnalysis objects from the 'analyses' attribute
            single_word_analyses_list = word_analysis_obj.analyses

            # Get the first SingleWordAnalysis object from this list
            first_analysis = single_word_analyses_list[0]

            # The SingleWordAnalysis object has a getLemmas() method.
            current_token_lemmas = first_analysis.getLemmas()

            if current_token_lemmas:
                lemmas.append(current_token_lemmas[0]) # Take the first lemma if available
            else:
                lemmas.append(token) # Fallback to original token if no lemmas are found
        else:
            lemmas.append(token) # Fallback if no WordAnalysis or no analyses found within it
    return lemmas

# Apply lemmatization to the 'tokens_stop' column
# Ensure 'tokens_stop' is available and correctly formatted (list of strings)
if 'tokens_stop' in df.columns:
    df['tokens_lemma_zemberek'] = df['tokens_stop'].apply(apply_zemberek_lemmatization)
    print("Lemmatisation Zemberek appliquée avec succès. Affichage des premières lignes:")
    display(df[['tokens_stop', 'tokens_lemma_zemberek']].head())
else:
    print("La colonne 'tokens_stop' n'a pas été trouvée. Veuillez vous assurer que les étapes de prétraitement précédentes ont été exécutées.")
    df['tokens_lemma_zemberek'] = None

# Update the df_lemmatized DataFrame with Zemberek results
df_lemmatized = df[["Musteri_ID", "tokens_lemma_zemberek"]].copy()
df_lemmatized.to_csv("lemmatized_veri_zemberek.csv", index=False)
print("DataFrame 'lemmatized_veri_zemberek.csv' a été sauvegardé.")
```
